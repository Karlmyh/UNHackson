```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
import csv
import numpy as np
```

```python
dutation_closure=pd.read_csv("./data/covid-9_duration_school_closures.csv")
date_closure=pd.read_csv("./data/covid-19-school-closure.csv")

```

```python
dutation_closure
date_closure
```

```python
panel_date_closure=pd.DataFrame([])
for i in range(210):
    country=dutation_closure["Country"].drop_duplicates().loc[i]
    panel_date_closure.index=date_closure[date_closure["Country"]==country].index
    panel_date_closure.loc[:,country]=date_closure[date_closure["Country"]==country]["Status"]
    
panel_date_closure.index=date_closure["Date"].drop_duplicates()
```

```python
panel_date_closure
```

```python

panel_date_closure[panel_date_closure=="Fully open"]=2
panel_date_closure[panel_date_closure=="Partially open"]=1
panel_date_closure[panel_date_closure=="Closed due to COVID-19"]=0
panel_date_closure[panel_date_closure=="Academic break"]=-1
panel_date_closure=panel_date_closure.drop(columns=["Svalbard"])
```

```python
panel_date_closure.to_csv("./cleaned_data/panel_date_closure.csv")

```

```python
cases=pd.read_csv("./data/WHO-COVID-19-global-data.csv")
```

```python
cases_cleaned=pd.DataFrame([])
deaths_cleaned=pd.DataFrame([])
for i in range(237):
    #print(cases["Country"].drop_duplicates())
    country=cases["Country"].drop_duplicates().iloc[i]
    cases_cleaned.index=cases[cases["Country"]==country].index
    deaths_cleaned.index=cases[cases["Country"]==country].index
    cases_cleaned.loc[:,country]=cases[cases["Country"]==country]["New_cases"]
    deaths_cleaned.loc[:,country]=cases[cases["Country"]==country]["New_deaths"]
    
cases_cleaned.index=cases["Date_reported"].drop_duplicates()
deaths_cleaned.index=cases["Date_reported"].drop_duplicates()
```
