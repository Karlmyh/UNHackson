```python
def average_data(data,flag=0):
    if flag==0:
        result=data.copy()
    
        for i in range(result.shape[0]):
                result[i]=data[i:min(result.shape[0],i+7)].mean()
    else:
        result=data.copy()

        for i in range(result.shape[0]):
            for j in range(result.shape[1]):
                result[i,j]=data[i:min(result.shape[0],i+7),j].mean()

    return result
```

```python

from scipy import stats


result=np.zeros([4,8])
cases_data=cases[44:-33].to_numpy()
school_shut_data=school_shut.to_numpy()
deaths_data=deaths[44:-33].to_numpy()

#cases_data=cases_data[60:240,:]
#school_shut_data=school_shut_data[60:240,:]
#deaths_data=deaths_data[60:240,:]

i=26
result[0,0]=cases_data[school_shut_data[:,i]==2,i].mean()
result[0,1]=cases_data[school_shut_data[:,i]==2,i].std()
result[0,2]=cases_data[school_shut_data[:,i]==1,i].mean()
result[0,3]=cases_data[school_shut_data[:,i]==1,i].std()
result[0,4]=cases_data[school_shut_data[:,i]==0,i].mean()
result[0,5]=cases_data[school_shut_data[:,i]==0,i].std()
result[0,6]=cases_data[school_shut_data[:,i]==-1,i].mean()
result[0,7]=cases_data[school_shut_data[:,i]==-1,i].std()

result[1,0]=deaths_data[school_shut_data[:,i]==2,i].mean()
result[1,1]=deaths_data[school_shut_data[:,i]==2,i].std()
result[1,2]=deaths_data[school_shut_data[:,i]==1,i].mean()
result[1,3]=deaths_data[school_shut_data[:,i]==1,i].std()
result[1,4]=deaths_data[school_shut_data[:,i]==0,i].mean()
result[1,5]=deaths_data[school_shut_data[:,i]==0,i].std()
result[1,6]=deaths_data[school_shut_data[:,i]==-1,i].mean()
result[1,7]=deaths_data[school_shut_data[:,i]==-1,i].std()


result[2,0]=result[0,0]-result[0,0]
result[2,2]=result[0,2]-result[0,0]
result[2,4]=result[0,4]-result[0,0]
result[2,6]=result[0,6]-result[0,0]

result[2,3]=stats.ttest_ind(cases_data[school_shut_data[:,i]==2,i],
                            cases_data[school_shut_data[:,i]==1,i]).pvalue<0.05
result[2,5]=stats.ttest_ind(cases_data[school_shut_data[:,i]==2,i],
                            cases_data[school_shut_data[:,i]==0,i]).pvalue<0.05
result[2,7]=stats.ttest_ind(cases_data[school_shut_data[:,i]==2,i],
                            cases_data[school_shut_data[:,i]==-1,i]).pvalue<0.05

result[3,0]=result[1,0]-result[1,0]
result[3,2]=result[1,2]-result[1,0]
result[3,4]=result[1,4]-result[1,0]
result[3,6]=result[1,6]-result[1,0]

result[3,3]=stats.ttest_ind(deaths_data[school_shut_data[:,i]==2,i],
                            deaths_data[school_shut_data[:,i]==1,i]).pvalue<0.05
result[3,5]=stats.ttest_ind(deaths_data[school_shut_data[:,i]==2,i],
                            deaths_data[school_shut_data[:,i]==0,i]).pvalue<0.05
result[3,7]=stats.ttest_ind(deaths_data[school_shut_data[:,i]==2,i],
                            deaths_data[school_shut_data[:,i]==-1,i]).pvalue<0.05





```
